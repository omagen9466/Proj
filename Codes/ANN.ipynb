```python
import pandas as pd
import numpy as np
df=pd.read_csv('Data/r.csv')

X=df.values[:,1:3]
X_means=np.mean(X,axis=0)
X_stds=np.std(X,axis=0)

X[:,0]=(X[:,0]-X_means[0])/X_stds[0]
X[:,1]=(X[:,1]-X_means[1])/X_stds[1]

df['$r$ [m]']=(df['$r$ [m]']-df['$r$ [m]'].mean())/df['$r$ [m]'].std()
# print(df['$r$ [m]'])
decimals = 3 
df['$r$ [m]'] = df['$r$ [m]'].apply(lambda x: round(x, decimals))
y=np.array(df['$r$ [m]'])
y_norm = (y - y.min()) / (y.max() - y.min())
print(y_norm)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

def create_model():
    
    model = Sequential()
    model.add(Dense(units=300, activation='relu', input_dim=2))
    model.add(Dense(units=300, activation='relu'))
    
    model.add(Dense(units=131, activation='linear'))
    
    optimizer = optimizers.Adam(learning_rate=0.001)
    model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['mse'])
    
    return model
```

```python
model = create_model()
history = model.fit(X, y_norm, epochs=500, batch_size=32)
y_pred = model.predict(X)
```
